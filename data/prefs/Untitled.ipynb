```python
import pandas as pd
```

```python
df = pd.read_csv('YAP1_prefs.csv.gz')
print(df['hgvs_pro'].str.extractall('(\d+)')[0].astype(int).max(level=0).max(),
     df['hgvs_pro'].str.extractall('(\d+)')[0].astype(int).min(level=0).min())
```
